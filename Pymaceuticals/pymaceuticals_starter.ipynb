```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_table = pd.merge(mouse_drug_data,clinical_trial_data,on='Mouse ID',how = 'outer')

# Display the data table for preview
merge_table.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Drug_Time = merge_table.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
Drug_n_Time = pd.DataFrame(Drug_Time)
Drug_n_Time.reset_index(inplace=True)
# Preview DataFrame
Drug_n_Time
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
hori_group = merge_table.groupby(['Drug','Timepoint'])

hori_group_df = pd.DataFrame(hori_group["Tumor Volume (mm3)"].sem())

hori_group_df.reset_index(inplace=True)

hori_group_df.head()
```

```python
table_df = Drug_n_Time.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
table_df.head()
```

```python
stdev = hori_group_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
plotStyleDict = {"Capomulin":'ro:', 'Infubinol':'b^:', 'Ketapril':'gs:', 'Placebo': 'kd:'}
fig1, ax1 = plt.subplots()
for drug in drugs:
     ax1.errorbar(table_df.index, table_df[drug], stdev[drug], fmt = plotStyleDict[drug])
ax1.legend(drugs)
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

plt.savefig("TumorResponseToTreatment.png")
plt.show()
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_response = merge_table.groupby(['Drug','Timepoint']).mean()
meta_response
# Convert to DataFrame
del meta_response["Tumor Volume (mm3)"]
# Preview DataFrame
meta_response.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
stdev_meta_response = merge_table.groupby(["Drug","Timepoint"]).sem()
# Convert to DataFrame
stdev_meta_response
# Preview DataFrame
del stdev_meta_response["Tumor Volume (mm3)"]
del stdev_meta_response["Mouse ID"]
stdev_meta_response = pd.DataFrame(stdev_meta_response)
stdev_meta_response.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
meta_response= pd.DataFrame(meta_response)
meta_response.reset_index(inplace=True)
new_pivot = meta_response.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
new_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
stdev_meta_response= pd.DataFrame(stdev_meta_response)
stdev_meta_response.reset_index(inplace=True)

stdev_meta_pivot = stdev_meta_response.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

plotStyleDict = {"Capomulin":'ro:', 'Infubinol':'b^:', 'Ketapril':'gs:', 'Placebo': 'kd:'}
fig2, ax2 = plt.subplots()
for drug in drugs:
    ax2.errorbar(new_pivot.index, new_pivot[drug], stdev_meta_pivot[drug], fmt = plotStyleDict[drug])
ax2.legend(drugs)
plt.title('Metastatic Spread during Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Metastatic Sites')
# Save the Figure
plt.savefig("MetastaticSpreadDuringTreatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survive = merge_table.groupby(["Drug","Timepoint"])
survive_rate = survive["Mouse ID"].count()
# Convert to DataFrame
survive_rate_df = pd.DataFrame(survive_rate)
# Preview DataFrame
survive_rate_df.reset_index(inplace = True)
survive_rate_df = survive_rate_df.rename(columns = {"Mouse ID":"Mouse Count"})
survive_rate_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survive_pivot = survive_rate_df.pivot(index='Timepoint', columns='Drug', values='Mouse Count')

# Preview the Data Frame
survive_pivot = pd.DataFrame(survive_pivot)
survive_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
for col in survive_pivot.columns:
    survive_pivot[col] = 100 * survive_pivot[col]/25
fig3, ax3 = plt.subplots()
for drug in drugs:
    plt.plot(survive_pivot.index, survive_pivot[drug], plotStyleDict[drug])
plt.grid(True)
ax3.legend(drugs)
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# Save the Figure
plt.savefig("SurvivalDuringTreatment.png")
# Show the Figure
plt.show()
```
